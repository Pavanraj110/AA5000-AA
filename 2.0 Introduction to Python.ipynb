# Python Beginners Tutorial

 ### We'll begin with basic variables.

Variables can be thought of as containers that store information, which can range from numerical values to text.

```python
# Storing a number
a = 73
print(a)
```

```python
# Or text
b = "This is some text"
print(b)
```

## What can we do with variables?

```python
# Let's add them together
a = 73
c = 27
d = a + c
print(d)
```

```python
# Or if you want to do it all in one line!
print(a+c)
```

```python
# We change variable values?
a = 37
print(a)
```

```python
# Copying variables
e = a
print(e)
```

```python
# Types of Varaibles

# Integer variable
a = 1
print(a)
print(type(a))

# Float variable (Non integer digits)
a = 5.0
print(a)
print(type(a))

# Boolean variable (True or False)
a = True
print(a)
print(type(a))

# Text variable also known as string variables
a = "7"
print(a)
print(type(a))
```

```python
# Be careful with variable operations

x = 10
y = 20
z = x + y
print(z)
print(type(z))

# string variables
x = "10"
y = "20"
z = x + y
print(z)
print(type(z))
```

# Array or List Variables
###  Arrays or lists are incredibly valuable in programming. Imagine them as containers with individual compartments, where each compartment can store another variable.

```python
# A Simple array or integers
a = [1,2,3]
print(a)
```

```python
# Storing variables of different types is possible
b =  ["Text", 4, 2.5]
print(b)
```

```python
# Accessing specific variables/locations in the list

# How to print the first element?
print(a[0])
print(a[1])
print(a[2])
# Note lists always start at index or position 0, so a list that holds 3 numbers can be indexed with 0,1 & 2
```

```python
# Storing more elements onto your list using *apend*

a.append(4)
print(a)
```

```python
# Getting our list length
print(len(a))
```

```python
# Changing the value of an element in a list

a[3] = 5
print(a)
```

```python
# Slicing lists is a technique used to get elements from, before or between different indices

print("All elelemts: ", a[:])
print("All elements from the 3rd position onward: ", a[2:])
print("All elements between the 2nd and 4th position: ", a[1:3])
print("All elements before the last postion", a[:-1])

# Very useful guide:

#a[start:stop]  # items start through stop-1
#a[start:]      # items start through the rest of the array
#a[:stop]       # items from the beginning through stop-1
#a[:]           # a copy of the whole array
```
