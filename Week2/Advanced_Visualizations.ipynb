# **Advanced Data visualization techniques for EDA**
- Violin
- Heatmaps
- Factor plots
- Density plots
- Joint Distribution plots

```python
import pandas as pd
import numpy as np

file_name = "https://raw.githubusercontent.com/Pavanraj110/AA5000-AA/main/Datasets/winequalityN.csv?token=GHSAT0AAAAAACN2CHF5FQ3KQMFWD2EAJLA4ZOLDRKA"
df = pd.read_csv(file_name)
df.head()
```

# Violin Plots

```python
import seaborn as sns
sns.violinplot(data=df['alcohol'])
```

# Density Plot


```python
# Assuming df is your DataFrame and it has columns named 'alcohol' and 'density'
sns.kdeplot(x=df['alcohol'], y=df['density'])
```

# Joint Distribution Plot

```python
sns.jointplot(x='alcohol', y='density', data=df)
```

# Factor Plots and Bee Swarm Plots

Factor plots allow us to create separate plots by categorical classes.

```python
# Factor Plot
g = sns.catplot(x='free sulfur dioxide',
                   y='alcohol',
                   data=df,
                   # A recent update in Matplotlib broke how the hue argument works in Seaborn with some plots see: https://github.com/mwaskom/seaborn/issues/1697
                   #hue='quality',  # Color by stage
                   col='quality',  # Separate by stage
                   kind='point') # Swarmplot

# Rotate x-axis labels
g.set_xticklabels(rotation=-45)
```

# Creating Empirical cumulative distribution function (ECDF)


```python
import numpy as np
import matplotlib.pyplot as plt
x = np.sort(df['alcohol'])
y = np.arange(1, len(x)+1) / len(x)
_ = plt.plot(x, y, marker='.', linestyle='none')
plt.margins(0.05)
_ = plt.xlabel('Percent of Alcohol in Wine')
_ = plt.ylabel('ECDF')
plt.grid(True)
plt.show()
```
