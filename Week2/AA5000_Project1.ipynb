```python
import pandas as pd

# Load the dataset
file_path = 'processed.cleveland.data'
column_names = [
    'age', 'sex', 'cp', 'trestbps', 'chol', 'fbs', 'restecg',
    'thalach', 'exang', 'oldpeak', 'slope', 'ca', 'thal', 'num'
]

df = pd.read_csv(file_path, names=column_names)
```

The dataset consists of 303 entries and 14 columns. Each column represents a different attribute related to heart disease diagnostics, including age, sex, chest pain type (cp), resting blood pressure (trestbps), cholesterol levels (chol), fasting blood sugar (fbs), resting electrocardiographic results (restecg), maximum heart rate achieved (thalach), exercise induced angina (exang), ST depression induced by exercise relative to rest (oldpeak), the slope of the peak exercise ST segment (slope), number of major vessels colored by fluoroscopy (ca), thalassemia (thal), and the diagnosis of heart disease (num). It's notable that the ca and thal columns are object types, possibly due to missing or non-numeric values.

```python
# Summary statistics for numerical columns
summary_statistics = df.describe()

# Checking for unique values in 'ca' and 'thal' to understand the object datatype
unique_ca = df['ca'].unique()
unique_thal = df['thal'].unique()

summary_statistics, unique_ca, unique_thal
```

The summary statistics provide an overview of the numerical variables in the dataset:

- Age: Ranges from 29 to 77 years, with a mean of approximately 54 years.

- Sex: Primarily binary (0 or 1), indicating female or male, respectively.

- Chest Pain Type (cp): Categorical, ranging from 1 to 4.

- Resting Blood Pressure (trestbps), Cholesterol (chol), Fasting Blood Sugar (fbs), Resting Electrocardiographic Results (restecg), Maximum Heart Rate Achieved (thalach), Exercise Induced Angina (exang), ST Depression Induced by Exercise Relative to Rest (oldpeak), and Slope of the Peak Exercise ST Segment (slope): These columns provide various physiological and clinical data points, each with its specific range and distribution.

- Diagnosis of Heart Disease (num): The target variable, indicating the presence of heart disease, with values from 0 to 4.

- For the columns ca (number of major vessels colored by fluoroscopy) and thal (thalassemia), which are of object type due to non-numeric values, we have identified the presence of '?' indicating missing or invalid entries.

#**EDA** - Visualization, cleaning and preperation.

```python
#Importing necessary libraries
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

##Transforming the data

```python
# Replacing '?' with NaN
df.replace('?', np.nan, inplace=True)

# Converting 'ca' and 'thal' to numeric
df['ca'] = pd.to_numeric(df['ca'])
df['thal'] = pd.to_numeric(df['thal'])

```

```python
# Checking the number of missing values in each column
missing_values = df.isna().sum()

# Imputation strategy: For simplicity, we'll impute missing values with the median of each column.
# However, a more sophisticated approach could be considered for a real-world scenario.
ca_median = df['ca'].median()
thal_median = df['thal'].median()

df['ca'].fillna(ca_median, inplace=True)
df['thal'].fillna(thal_median, inplace=True)

# Re-checking the number of missing values after imputation
missing_values_post_imputation = df.isna().sum()

missing_values, missing_values_post_imputation
```

##Visualzing the dataset

```python
# Basic EDA
# Plotting distributions of key numerical variables
fig, axes = plt.subplots(2, 3, figsize=(18, 12))
sns.histplot(df['age'], bins=20, kde=True, ax=axes[0, 0]).set(title='Age Distribution')
sns.histplot(df['chol'], bins=20, kde=True, ax=axes[0, 1]).set(title='Cholesterol Distribution')
sns.histplot(df['trestbps'], bins=20, kde=True, ax=axes[0, 2]).set(title='Resting Blood Pressure Distribution')
sns.histplot(df['thalach'], bins=20, kde=True, ax=axes[1, 0]).set(title='Max Heart Rate Distribution')
sns.countplot(x='sex', data=df, ax=axes[1, 1]).set(title='Sex Distribution')
sns.countplot(x='cp', data=df, ax=axes[1, 2]).set(title='Chest Pain Type Distribution')

plt.tight_layout()

# Checking the distribution of the target variable 'num'
plt.figure(figsize=(8, 4))
sns.countplot(x='num', data=df).set(title='Distribution of Heart Disease Diagnosis')
plt.show()
```

The exploratory data analysis (EDA) provides valuable insights into the dataset:

- Age Distribution: The age of participants varies widely, with a concentration between 50 and 60 years, indicating a middle-aged to older population.

- Cholesterol Distribution: Cholesterol levels show a broad distribution, with a peak around 200-250 mg/dl.

- Resting Blood Pressure Distribution: Resting blood pressure levels are mostly clustered around 120-140 mmHg.

- Max Heart Rate Distribution: The maximum heart rate achieved during the test varies significantly among participants, with a concentration around 150-170 bpm.

- Sex Distribution: The dataset includes more male participants than female.

- Chest Pain Type Distribution: The most common type of chest pain experienced by participants is type 4 (asymptomatic), followed by type 3 (non-anginal pain), type 2 (atypical angina), and type 1 (typical angina).

The distribution of the target variable, Heart Disease Diagnosis (num), shows a variety of outcomes, with a significant number of participants having no diagnosis of heart disease (num=0) and others showing different degrees of disease.

This analysis provides a good starting point for further investigation, such as examining correlations between variables, performing hypothesis testing, or preparing the data for predictive modeling. Handling missing values in ca and thal is crucial before any advanced analysis or modeling. One could impute these missing values based on the distribution of the data or by using more sophisticated techniques like modeling the missing data based on other variables.

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Calculating the correlation matrix
correlation_matrix = df.corr()

# Plotting the correlation matrix as a heatmap
plt.figure(figsize=(14, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix of Heart Disease Dataset')
plt.show()
```

- **Positive Correlations:** Variables like age, sex, cp (chest pain type), trestbps (resting blood pressure), chol (cholesterol), and oldpeak (ST depression induced by exercise) show positive correlations with the target variable num. This indicates that higher values of these variables are associated with a greater likelihood of heart disease.

- **Negative Correlations:** Variables such as thalach (maximum heart rate achieved) show a negative correlation with num, suggesting that higher maximum heart rates are associated with a lower likelihood of heart disease.

- **Weak Correlations:** Some variables like fbs (fasting blood sugar) and restecg (resting electrocardiographic results) show weaker correlations with the target variable, indicating they might be less predictive of heart disease on their own.

```python
# Visual exploration of relationships with the target variable 'num'
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

# Chest Pain Type vs Heart Disease
sns.countplot(x='cp', hue='num', data=df, ax=axes[0], palette='Set1')
axes[0].set_title('Chest Pain Type vs Heart Disease')
axes[0].set_xlabel('Chest Pain Type')
axes[0].set_ylabel('Count')

# Maximum Heart Rate vs Heart Disease
sns.boxplot(x='num', y='thalach', data=df, ax=axes[1], palette='Set2')
axes[1].set_title('Max Heart Rate Achieved vs Heart Disease')
axes[1].set_xlabel('Heart Disease Diagnosis')
axes[1].set_ylabel('Max Heart Rate')

# ST Depression vs Heart Disease
sns.boxplot(x='num', y='oldpeak', data=df, ax=axes[2], palette='Set3')
axes[2].set_title('ST Depression Induced by Exercise vs Heart Disease')
axes[2].set_xlabel('Heart Disease Diagnosis')
axes[2].set_ylabel('ST Depression')

plt.tight_layout()
plt.show()
```

The visual exploration of relationships between key variables and the presence of heart disease reveals some interesting patterns:

**Chest Pain Type vs. Heart Disease:** Different types of chest pain (cp) show varying distributions across heart disease diagnoses. It appears that certain types of chest pain are more frequently associated with higher levels of heart disease diagnosis.

**Max Heart Rate Achieved vs. Heart Disease:** The boxplot indicates that individuals with a higher maximum heart rate (thalach) tend to have a lower diagnosis level of heart disease. This suggests a potential inverse relationship between maximum heart rate and the severity of heart disease.

**ST Depression Induced by Exercise vs. Heart Disease:** There is a visible trend where higher levels of ST depression (oldpeak) are associated with higher levels of heart disease diagnosis. This suggests that oldpeak could be a significant predictor of heart disease presence.

These visualizations, along with the correlation analysis, can help in identifying key variables that might be significant predictors in the context of heart disease diagnosis. Such insights are valuable for developing predictive models or for clinical decision-making.
