##Research Question.
**"How do satisfaction level, last evaluation, and number of projects linearly relate to the probability of an employee leaving the company?"**


```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix

# Selecting features and target variable
X = df[['satisfaction_level', 'last_evaluation', 'number_project']]
y = df['left']

# Splitting the data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

# Initiating and fitting the linear regression model
lin_reg = LinearRegression()
lin_reg.fit(X_train, y_train)

# Predicting on the test set
y_pred_lin = lin_reg.predict(X_test)

# Calculating the Mean Squared Error (MSE) and R-squared value
mse = mean_squared_error(y_test, y_pred_lin)
r2 = r2_score(y_test, y_pred_lin)

mse, r2
# Applying a threshold of 0.5 to classify predictions
y_pred_classified = [1 if pred > 0.5 else 0 for pred in y_pred_lin]

# Calculating accuracy
accuracy = (y_test == y_pred_classified).mean()

accuracy
print('mean squared error:',mse)
print('R squared value:',r2)
print('Accuracy',accuracy)
```

To adapt our analysis to a linear regression model, we need to revise our approach slightly, considering that the outcome variable (`left`) is binary (0 for stayed, 1 for left). While linear regression is typically used for continuous outcome variables, we can still apply it here for educational purposes to see if there's a linear relationship between the predictors (such as `satisfaction_level`, `last_evaluation`, `number_project`) and the likelihood of an employee leaving, interpreted through the continuous scale of probabilities.

### Implementing Linear Regression Model:

### Linear Regression Model Results

The linear regression model, fitted with the predictors `satisfaction_level`, `last_evaluation`, and `number_project`, produced the following metrics for evaluating its performance:

- **Mean Squared Error (MSE):** 0.154
- **R-squared Value:** 0.164

### Interpretation:

- **MSE (Mean Squared Error):** This value quantifies the average squared difference between the actual binary outcomes (0 or 1 for employees who stayed or left, respectively) and the predicted values (continuous scale). A lower MSE indicates a closer fit of the model to the data. In this context, an MSE of 0.154 suggests that, on average, the squared difference between the predicted values and the actual values is relatively small, considering the outcome variable is binary.
  
- **R-squared Value:** The R-squared value, or the coefficient of determination, explains the proportion of the variance in the dependent variable that is predictable from the independent variables. An R-squared value of 0.164 means that approximately 16.4% of the variation in whether employees leave or stay can be explained by the model's predictors (satisfaction level, last evaluation, and number of projects). While this is relatively low, it indicates that these factors do have some linear relationship with the likelihood of an employee leaving, but other factors not included in the model might also play significant roles.

### Conclusion:

Using a linear regression model to predict a binary outcome has provided some insight into the relationship between employee satisfaction, evaluations, project load, and their departure from the company. However, the relatively low R-squared value suggests that the predictors included in the model do not fully capture the complexity of the decision-making process behind employee turnover.

This analysis underscores the importance of considering the appropriate model type for the prediction task at hand. For binary outcomes like employee turnover, logistic regression or other classification methods are generally more suitable and provide more interpretable results in terms of probabilities.
